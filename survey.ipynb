```r
survey <- read.csv(file="W3 Project Student Entry Survey.csv")
```

```r
devtools::install_github('mattroumaya/doubleheadr')
```

```r
library(doubleheadr)
```

```r
colnames(clean_headr(survey,"X."))
```

```r
result <- trim_headr(clean_headr(survey,"X."),c(
    "how_confident_are_you_that_you_could_do_each_of_the_following_",
    "what_is_your_race_mark_all_that_apply_",
    "the_next_few_questions_are_about_relationships_and_dating_please_answer_the_questions_below_even_if_you_are_not_currently_dating_or_going_out_with_someone_11_for_each_of_the_items_below_please_mark_how_true_each_statement_is_of_you_",
    "for_each_of_the_items_below_please_mark_how_true_each_statement_is_of_you_",
    "how_confident_are_you_that_you_could",
    "_caregiver_or_another_trusted_adult_about_each_of_the_following_"
    ))
colnames(result)
```

```r
write.csv(result,"result.csv")
```
